<h2>Converting genres to a list</h2>

```python
movies.genres[1]
```

```python
import ast
```

```python
def convert_genres_to_list(obj):
    genres_list = []
    for i in ast.literal_eval(obj):
        genres_list.append(i['name'])
    return genres_list
```

```python
convert_genres_to_list(movies.genres[1])
```

```python
movies.genres.apply(convert_genres_to_list)
```

```python
movies['genres'] = movies.genres.apply(convert_genres_to_list)
```

```python
movies.head()
```

```python
#Convert keywords to list
movies['keywords'] = movies.keywords.apply(convert_genres_to_list)
```

```python
movies.head(1)
```

```python
print(movies['cast'][0])
```

```python
#We need to take the name of first 3 cast
def convert_cast_to_list(obj):
    cast_list = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter < 3:
            cast_list.append(i['name'])
            counter += 1
    return cast_list
```

```python
convert_cast_to_list(movies['cast'][0])
```

```python
movies.cast = movies.cast.apply(convert_cast_to_list)
```

```python
movies.crew[1]
```

```python
# movies_crew_list = ast.literal_eval(movies.crew[1])
# movies_crew_list

def fetch_director_name_from_crew(obj):
    director_list = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            director_list.append(i['name'])
    return director_list
```

```python
fetch_director_name_from_crew(movies.crew[1])
```

```python
movies.crew = movies.crew.apply(fetch_director_name_from_crew)
```

```python
#converting tags to its root words
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem("loving")
```

```python
def convert_to_root_words(text):
    word_list = []
    for i in text.split():
        word_list.append(ps.stem(i))
    return " ".join(word_list)
```

```python
convert_to_root_words('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
movies_df.tags = movies_df.tags.apply(convert_to_root_words)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(movies_df['tags']).toarray()
```
